# Dummy data

```python
dummy_data = [[1, 2, 3, 10, 11, 12], [40, 50, 60, 130, 140, 100]]

X = pd.DataFrame(data=dummy_data).T.rename(columns={0: "Var1", 1: "Var2"})
X
```

```python
scaler.fit(X)
```

# Nuestro propio Transformer

```python
from sklearn.base import TransformerMixin, BaseEstimator
```

```python
class MyStandardScaler(BaseEstimator, TransformerMixin):

    def __init__(self, verbose):
        self.verbose = verbose

    def fit(self, X):

        mean = X.mean()
        var = X.var(ddof=0)

        if self.verbose:
            print(mean)
            print(var)

        self.mean_ = mean
        self.var_ = var

        return self

    def transform(self, X):

        Xt = (X - self.mean_) / (self.var_**0.5)

        return Xt
```

```python
my_scaler = MyStandardScaler(verbose=False)
```

```python
my_scaler.fit(X)
```

```python
my_scaler.mean_
```

```python
my_scaler.var_
```

```python
Xt_myscaler = my_scaler.transform(X)
```

```python
Xt_myscaler
```

```python
pd.concat(
    [
        Xt_scaler.describe().round(2),
        Xt_pandas.describe().round(2),
        Xt_myscaler.describe().round(2),
    ],
    axis=1,
)
```
